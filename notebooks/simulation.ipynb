## Simulation

We extract the simulation results here and create the plots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
names = {
    'constat': 'ConStat',
    'normal': 'Normalized-Test',
    'mean': 'Mean-Test',
    'constat_no_bootstrap': 'ConStat-No-Bootstrap',
    'constat_no_sort': 'ConStat-No-Sort',
    'constat_no_random': 'ConStat-No-Random',
}

def plot(df, methods, save_path=None, add_linear_fit=True):
    p_values = df[methods]
    fig, ax = plt.subplots()
    for method in p_values.columns:
        sns.ecdfplot(p_values[method], label=names[method])
    ax.set_xlabel(r'$p$', fontsize=21)
    ax.set_ylabel(r'$F_p$', fontsize=21, rotation=0, labelpad=10, y=0.45)
    if add_linear_fit:
        x = np.linspace(0.000000001, 1, 100)
        y = x
        ax.plot(x, y, 'k--', label='Identity')
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    # set grey background color
    ax.tick_params(axis='both', which='major', labelsize=21)
    ax.set_yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    ax.set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    ax.set_facecolor((0.95, 0.95, 0.95))
    ax.legend(fontsize=14)
    ax.set_xlim(-0.03, 1.03)
    ax.set_ylim(-0.03, 1.03)
    fig.tight_layout()
    if save_path is not None:
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        fig.savefig(save_path)
```

```python
df = pd.read_csv('../data/simulation/easy_scenario_uncontamination.csv')

plot(df, ['constat', 'mean', 'normal', 'constat_no_sort', 'constat_no_bootstrap', 'constat_no_random'])
```
